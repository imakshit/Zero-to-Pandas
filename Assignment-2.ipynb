```python
# Example 1 
x = np.array([510, 560, 5368, 536])

np.prod(x)
```

## Function 3 - ndArray.flatten()
Return a copy of the array collapsed into one dimension

```python
# Example 1 - working
arr = np.array([[1,2],[89,98],[45,68]])
arr.flatten()
```

returns a copy of arr in 1-dimensions

```python
# Example 2 - working
x = np.array([[2,3],[4,5]])
x.flatten()
```

returns a copy of x in 1-dimensions

```python
# Example 3 - breaking (to illustrate when it breaks)
# No case in which function breaks
# will only break if any syntax array or any argument missing
# I will explain some more concepts regarding this function here


#you can provide the optional argument 'order' to flatten the array in a specific order

# we will flatten in column-major
x.flatten('F')
```

order{‘C’, ‘F’, ‘A’, ‘K’}
‘C’ means to flatten in row-major (C-style) order. ‘F’ means to flatten in column-major (Fortran- style) order. ‘A’ means to flatten in column-major order if a is Fortran contiguous in memory, row-major order otherwise. ‘K’ means to flatten a in the order the elements occur in memory. The default is ‘C’.

we can use this function to create a copy of array in 1 dimensions

## Function 4 - np.transpose()
We can generate the transposition of an array using the tool numpy.transpose().
It will not affect the original array, but it will create a new array.

```python
# Example 1 - working

my_array = np.array([[1,2,3],
                        [4,5,6]])
print("transposed array : ")
print(np.transpose(my_array))
print("original array : ")
print(my_array)
```

returns transpose of array without affecting the original array

```python
# Example 2 - working
x = np.arange(9).reshape((3,3))
print("transposed array: ")
print(np.transpose(x))
print("original array: ")
print(x)
```

Explanation about example

```python
# Example 3 - breaking (to illustrate when it breaks)
# No case in which function breaks
# will only break if any syntax array or any argument missing
# I will explain some more concepts regarding this function here

x = np.ones((2,1,3))
print(x.shape)
y = np.transpose(x,(2,0,1)).shape
print(y)
```

Use transpose(a, argsort(axes)) to invert the transposition of tensors when using the axes keyword argument.

This function can be used to calculate transpose of a matrix

## Function 5 - np.eye()

The eye tool returns a 2-D array with 's as the diagonal and 's elsewhere. The diagonal can be main, upper or lower depending on the optional parameter . A positive  is for the upper diagonal, a negative  is for the lower, and a   (default) is for the main diagonal.

```python
# Example 1 - working
np.eye(8,7)
```

returns an 8x7 array with diagonal as 1

```python
# Example 2 - working
np.eye(8,8,k=2)
```

returns an array with diagonal as 1 starting from index 2

```python
# Example 3 - breaking (to illustrate when it breaks)
# No case in which function breaks
# will only break if any syntax array or any argument missing
# I will explain some more concepts regarding this function here

np.eye(8,8,10)
```

If the value of 'K' is out of range, the function will simply return an array of zeroes with given dimensions


This function can be used to generate inverse of identity array
